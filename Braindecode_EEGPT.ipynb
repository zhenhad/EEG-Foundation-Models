EEGPT turned your EEG window into a grid of 31 × 4 = 124 tokens, and each token is represented by a 512-dim vector.


What feats.shape = [1, 31, 4, 512] means

Think of EEGPT as producing a 512-dim feature vector for each token.

1 = batch size (one EEG window)

31 = number of time patches (your 1000 samples got split into 31 patches by the model’s patching setup)

4 = number of channel groups / channel patches (EEGPT doesn’t treat each channel as a separate token here; it groups channels into 4 spatial chunks)

512 = embedding dimension (the feature size per token)

```python
# Most downstream models want a single vector per example. The simplest is average pooling over the token grid:
# feats: (B, T_patches, C_groups, D)
window_embedding = feats.mean(dim=(1, 2))  # (B, D)

print("Window embedding shape:", window_embedding.shape)
```

This pipeline is now:

EEG window (62 × 1000) → EEGPT encoder → token embeddings (31 × 4 × 512) → mean pool → window embedding (512)


Freeze EEGPT, use it only as a feature extractor, and train a single linear layer on top of the 512-dim embeddings.

```python
# Step 1) Prepare a DataLoader:
# Iterate over windows, extract embeddings, and train a linear classifier.
# Build a PyTorch Dataset that returns embeddings + labels

import torch
from torch.utils.data import Dataset

label_map = {1: 0, 2: 1, 3: 2, 4: 3}

class EEGPTEmbeddingDataset(Dataset):
    def __init__(self, windows_list, embed_model):
        self.windows = windows_list
        self.embed_model = embed_model
        self.embed_model.eval()

    def __len__(self):
        return len(self.windows)

    def __getitem__(self, idx):
        X, y, _ = self.windows[idx]
        y = label_map[int(y)]

        X = torch.tensor(X[:62, :]).float()  # (62, 750)

        # pad time dimension from 750 -> 1000 (pad zeros at the end)
        if X.shape[1] < 1000:
            pad = 1000 - X.shape[1]
            X = torch.nn.functional.pad(X, (0, pad))  # (62, 1000)

        X = X.unsqueeze(0)  # (1, 62, 1000)

        with torch.no_grad():
            feats = self.embed_model(X)
            emb = feats.mean(dim=(1, 2)).squeeze(0)  # (512,)

        return emb, torch.tensor(y, dtype=torch.long)

# quick sanity check
train_ds = EEGPTEmbeddingDataset(train_windows, embed_model)
print(train_ds[0][0].shape, train_ds[0][1])

```
